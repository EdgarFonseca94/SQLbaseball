### From 2006 to 2016, which 5 teams had the highest payrolls?

```python
# highest paid teams of 2006 and 2016
payrolls = pd.read_sql("""
WITH payrolls AS
(SELECT SUM(salary) AS Payroll, teamID, yearID 
FROM Salaries
WHERE yearID BETWEEN 2006 AND 2016
GROUP BY teamID, yearID
ORDER BY SUM(salary) DESC
LIMIT 20)

SELECT DISTINCT(teamID) FROM payrolls
LIMIT 5;
""", db)
payrolls
```

From 2006 to 2016 these teams had the highest payrolls. The New York Angels, LA Dodgers, Detroit Lions, Boston Red Sox, and the Philidelphia Phillies.

### How do the top 5 Teams Payrolls compare to other teams?


```python
# used CTE to get top teams by payroll descending order
# then used case when to make column where if teamID not in the top 5 teams then it would be tagged as other

payrolls_5 = pd.read_sql("""
WITH payrolls AS
(SELECT SUM(salary) AS Payroll, teamID, yearID 
FROM Salaries
WHERE yearID BETWEEN 2006 AND 2016
GROUP BY teamID, yearID
ORDER BY SUM(salary) DESC)

SELECT Payroll,yearID, AVG(Payroll),
    CASE WHEN teamID NOT IN (SELECT DISTINCT(teamID) FROM payrolls LIMIT 5) THEN 'Others'
    ELSE teamID
    END team
FROM payrolls
GROUP BY Team, yearID
""", db)
payrolls_5
```

```python
# create visualization for this
(
    ggplot(payrolls_5, aes(x = "yearID", y = "Payroll", color = "team")) + geom_line() +
    labs(title = "Payrolls For Top 5 Teams vs Others ")
)
```

We can see that the top 5 teams Payrolls were higher than the average of 'Other' teams. Philly had their payrolls cut dramatically after 2014. The payrolls for LA Dodgers increased significanlty after 2012.

### Is there any general trend in salaries over this time period?

```python
# average payroll over time
avg_payrolls = pd.read_sql("""
SELECT yearID, AVG(salary) AS Payroll
FROM Salaries
WHERE yearID BETWEEN 2006 AND 2016
GROUP BY yearID
""", db)
avg_payrolls
```

```python
# graph avg payrolls over time
(
    ggplot(avg_payrolls, aes(x = "yearID", y = "Payroll")) + geom_line() + labs(title = "Average Salary 2006-2016")
)
```

We can see that average salaries increased overtime significantly.

### Highest paid athletes of all time?

```python
# query for highest paid athletes
highest = pd.read_sql("""
SELECT l.*, r.nameGiven,r.nameLast FROM Salaries l
LEFT JOIN People r
ON l.playerID = r.playerID
ORDER BY salary DESC
LIMIT 20
""", db)
highest
```

Here we can see that Alex Rodriguez had the highest salary of all time. Kershaw is also one of the players that had a high salary. Lets look at some of these well payed players and see how their salary progressed over time.

```python
# query to get unique players on list above to get their ids and see how their salaries progressed

# Made CTE to use in a join with Salaries to get only players that were in our highest paid athletes list

highest_salaries = pd.read_sql("""
WITH highest AS (
SELECT l.*, r.nameGiven,r.nameLast FROM Salaries l
LEFT JOIN People r
ON l.playerID = r.playerID
ORDER BY salary DESC
LIMIT 20)

SELECT * FROM Salaries
WHERE playerID IN (SELECT DISTINCT(playerID) FROM highest LIMIT 5);
""", db)
highest_salaries
```

```python
# Graphing the salaries of all these athletes which are 11.
# graph avg payrolls over time
(
    ggplot(highest_salaries, aes(x = "yearID", y = "salary", color = 'playerID')) + 
    geom_line() + labs(title = "Highest Paid Athletes Salaries Overtime")
)
```

We can see that Alex Rodriguez had highest paid salary in 2009, but then declined after 2010. All other players had very high upward salary progression.

### Top Players by Homeruns

```python
# query to get top 10 players by the sum of homeruns
pd.read_sql("""
SELECT SUM(HR) AS TotHR,
playerID FROM Batting
GROUP BY playerID
ORDER BY TotHR DESC LIMIT 10
""", db)
```

```python
# query to get salaries of top batters
top_batters = pd.read_sql("""
SELECT l.*, r.salary FROM
    (SELECT playerID, HR, yearID FROM Batting
    WHERE playerID IN ('bondsba01' ,'rodrial01', 'pujolal01')) as l
LEFT JOIN
    (SELECT * FROM Salaries) as r
ON l.playerID = r.playerID AND l.yearID = r.yearID
""", db)
top_batters
```

```python
(
    ggplot(top_batters, aes(x = "yearID", y = "HR", color = "playerID")) + geom_line() + facet_wrap("playerID", scales = 'free')+
    labs(title = "# Of Homeruns Throughout Career")
)
```

From the number of homeruns by year graphs above we can see that all 3 players number of homeruns increased, but then eventually declined maybe towards the end.

```python
(
    ggplot(top_batters, aes(x = "yearID", y = "salary", color = "playerID")) + geom_line() + facet_wrap("playerID", scales = 'free')+
    labs(title = "Salary Throughout Career")+ theme(subplots_adjust={'wspace': .1})
)
```

I wanted to see how the number of homeruns affected salaries. I decided on three players that I knew about because they were more modern players. The three players were Barry Bonds, Alex Rodriguez, and Albert Pujols. In the two graphs above we can see the lineplots for the number of homeruns and salaries. ### How Win Percentage Affects Attendance

```python
# query to get attendance numbers and win percentage
attendance = pd.read_sql("""
SELECT l.*, r.attendance FROM
    (SELECT yearID, W, L, G, ((W * 1.0)/(G * 1.0)) AS WinPct, name, TeamID FROM Teams
WHERE yearID >= 1980) AS l
LEFT JOIN (SELECT * FROM Homegames WHERE year_key >= 1980) as r
ON l.yearID = r.year_key AND l.teamID = r.team_key;
""", db)
```

```python
# graph win percentage vs attendance
(
    ggplot(attendance, aes(x = "WinPct", y = "attendance")) + geom_point()+
    labs(title = "Win Percentage vs Attendance")
)
```

My next question was if win percentage, which I calculated as the number of wins over total games played, affected the attendance in home games. I expected to see what our scatterplot shows which is a positive relationship between these two, but there were some occurences where there was a very high percentage of win rate but very low attendance.

```python
pd.read_sql("""
WITH low_attendance AS(
SELECT * FROM
(SELECT l.*, r.attendance, r.park_key, r.games FROM
(SELECT yearID, W, L, G, ((W * 1.0)/(G * 1.0)) AS WinPct, name, TeamID FROM Teams
WHERE yearID >= 1980) AS l
LEFT JOIN
    (SELECT * FROM Homegames WHERE year_key >= 1980) as r
ON l.yearID = r.year_key AND l.teamID = r.team_key)
WHERE attendance < 50000)

SELECT l.*, r.park_name FROM low_attendance l
LEFT JOIN (SELECT * FROM Parks) as r
ON l.park_key = r.park_key
ORDER BY WinPct DESC
            ;""", db)
```

I decided to look into why there was a very small attendance for some teams with high win percentage. I used the table from before but added the name of the field they played in and number of games played there. We can see that only a small number of games were played in these fields. So this led to my conclusion that these weren't the places they regularly played at. I looked further in the web to find some some clues of why they would play there and found this article of the Houston Astros in 2017. https://www.mlb.com/press-release/tropicana-field-to-host-three-game-rangers-vs-astros-series-starting-t-251115480 This article state the cause of the Houston Astros playing at Tropican Field was because of Hurricane Harvey which leads me to believe the other teams also may have had similar reasons.

### Total Number of Wins and World Series Wins

```python
## total number of wins from 2000-2018

TotWins= pd.read_sql("""
SELECT SUM(W) AS TotWins, SUM(L) as TotLoss,
teamID, name FROM Teams
WHERE yearID >= 2000
GROUP BY teamID
ORDER BY Totwins DESC
""", db)
```

```python
## total number of World series Wins

WSWIN = pd.read_sql("""
SELECT SUM(CASE WHEN WSWin = 'Y' THEN 1 ELSE 0 END) AS TotWSWin,
name FROM Teams
WHERE yearID >= 2000
GROUP BY name
ORDER BY TotWSWin DESC LIMIT 12
""", db)
```

```python
# Graphing wins to compare
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (10,7))
sns.barplot(x='TotWins', y="name", color = 'cornflowerblue' ,data=TotWins, order=TotWins['name'], ax = ax1)
sns.barplot(x='name', y="TotWSWin", color = 'cornflowerblue',data=WSWIN, order=WSWIN['name'], ax = ax2)

ax1.set_xlabel("TotWins")
ax1.set_title('Total Number of Wins 2000-2018')
ax2.set_title('Total Number of World Series Wins 2000-2018')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha = 'right')

plt.show()
```

For my last question I decided to see which team had the highest number of wins during 2000 to 2018. In our graph from the left we can see that the top 5 teams are Yankees, Cardinals, Red Sox, Dodgers, and the Braves. I then wanted to see if number of wins translated to the teams winning the World Series.

The graph in the right shows the number of times from 2000 to 2018 that the teams have won a World Series. We can see that teams with more than 1 win in the World Series are in our top 7 winners of games. There are also teams that won at least 1 World Series that don’t have high number of wins for example the Kansas City Royals. There is a team that has very high number of wins but did not win a World Series during this time, which is the Los Angeles Dodgers.
